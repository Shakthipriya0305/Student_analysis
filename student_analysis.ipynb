```python
# Data Science/Data Analysis with Python-Task 1
## Student Performance Analysis
# Name : Shakthipriya
# Date :

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
# Here the student performance data is loaded using pandas

df = pd.read_csv("/student-mat.csv",sep=';')
df.head()

# Data Exploration and Cleaning
# Various exploration techniques are implemented

# To check null values in a column
print(df.isnull())

# To check if there any null values in the dataset
print(df.isnull().values.any())

# To find the duplicate rows
print(df.duplicated())

# To remove the duplicate rows
df=df.drop_duplicates()

# Identify the dataset shape
print("Shape of thr dataset:",df.shape)

# Identify the datatype
print("\n Data Types")
print(df.dtypes)

# Averge of G3
average_grade = df['G3'].mean()
print("Average Score:",average_grade)
# The average is around 10.42 shows the overall performance of the students

# Student who scored above 15
high_scorers = (df['G3'] > 15).sum()
print("Students scoring above 15:", high_scorers)
# The total no. of students scored above 15 is 40

#Correlation between studytime and G3
correlation = df[['studytime','G3']].corr()
print(correlation)
# The correlation is 0.97 which is very strong positive correlation indicating the good study hours provides good result

# Performance based on Gender of the student
gender_average = df.groupby('sex')['G3'].mean()
print(gender_average)
# The Boys have 10.91 whereas the Girls have 9.96.Therefore Boys are performing slightly better the Girls

# Histogram of Grades
plt.figure(figsize=(8,5))
plt.hist(df['G3'],bins=6,edgecolor="orange")
plt.title("Histogram of Grades")
plt.xlabel("Final Grade G3")
plt.ylabel("Number of Students")

plt.show()
#From the histogram,most of the students scored between 10 - 12 in the final grade

#StudyTime Vs Grades
plt.figure(figsize=(8,5))
plt.scatter(df['studytime'],df['G3'])
plt.title("Study time Vs Grades")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")

plt.show()
#From this plot we can define that students with more study time tend to score highest score than students having less study time.
#Therefor this plot shows a positive performance analysis

# Average Final Grade by Gender
gender_avg = df.groupby('sex')['G3'].mean()
plt.figure(figsize=(8,5))
gender_avg.plot(kind="bar")
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")

plt.show()
# The bar graph shows that boys have scored slightly better than girls.

```
